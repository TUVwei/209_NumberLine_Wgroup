```python
import numpy as np
```

# 1 KF with no potential field

```python
ry = 1
fi = 0.1
qv = 1
H = 100
```

```python
s = 0
o = 0
m_pri = 0
v_pri = 1
for t in range(H):
    m_pos = m_pri + fi
    v_pos = v_pri + (t+1)*qv
    
    # update
    s = s + fi + np.random.randn() * np.sqrt((t+1)*qv)
    o = s + np.random.randn() * np.sqrt(ry)
    
    # next
    tmp_m_pri = m_pos + v_pos/(v_pos + ry)*(o - m_pos)
    tmp_v_pri = v_pos - v_pos/(v_pos + ry)*v_pos
    
    m_pri = tmp_m_pri
    v_pri = tmp_v_pri
    print(s, m_pri, v_pri)
```

```python
ry = 1
fi = 0.01
qv = 1
H = 100
```

```python
s = 0
o = 0
m_pri = 0
v_pri = 1
for t in range(H):
    m_pos = 3*m_pri + fi
    v_pos = 3*v_pri*3 + (t+1)*qv
    
    # update
    s = s + fi + 2*np.sin(s) + np.random.randn() * np.sqrt((t+1)*qv)
    o = s + np.random.randn() * np.sqrt(ry)
    
    # next
    tmp_m_pri = m_pos + v_pos/(v_pos + ry)*(o - m_pos)
    tmp_v_pri = v_pos - v_pos/(v_pos + ry)*v_pos
    
    m_pri = tmp_m_pri
    v_pri = tmp_v_pri
    print(s, m_pri, v_pri)
```
